<h1><b>Explorary Data Analysis<b></h1>


<h4><b>Gender</b></h4>


```python
df.columns
```

```python
# Count : Males vs Females
ax = sns.countplot(x="Gender", hue="Gender", data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(
    by="Amount", ascending=False
)
```

```python
# Males vs Females Purchasing Power
sales_gen = (
    df.groupby(["Gender"], as_index=False)["Amount"]
    .sum()
    .sort_values(by="Amount", ascending=False)
)
sns.barplot(x="Gender", y="Amount", data=sales_gen, hue="Gender")
```

<ul>
From above graphs we can conclude that :
<li>Most of the buyers are females</li>
<li>Purchasing power of females is greater than males</li>
</ul>


<h4><b>Age</b></h4>


```python
df.columns
```

```python
# Age Group GenderWise
ax = sns.countplot(data=df, x="Age Group", hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = (
    df.groupby(["Age Group"], as_index=False)["Amount"]
    .sum()
    .sort_values(by="Amount", ascending=False)
)
sns.barplot(x="Age Group", y="Amount", data=sales_age)
```

<ul>
From above graphs we can conclude that :
<li>Most of the buyers are of age group between 26-35 years females.</li>
</ul>


<h4><b>State</b></h4>


```python
df.columns
```

```python
# Total Orders from Top 10 States

sales_state = (
    df.groupby(["State"], as_index=False)["Orders"]
    .sum()
    .sort_values(by="Orders", ascending=False)
    .head(10)
)

sns.set(rc={"figure.figsize": (16, 5)})
sns.barplot(data=sales_state, x="State", y="Orders")
```

```python
# Total sales from Top 10 States

sales_state = (
    df.groupby(["State"], as_index=False)["Amount"]
    .sum()
    .sort_values(by="Amount", ascending=False)
    .head(10)
)

sns.set(rc={"figure.figsize": (16, 5)})
sns.barplot(data=sales_state, x="State", y="Amount")
```

<h4><b>Marital Status</b></h4>


```python
ax = sns.countplot(x="Marital_Status", data=df)
sns.set(rc={"figure.figsize": (6, 5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital = (
    df.groupby(["Marital_Status", "Gender"], as_index=False)["Amount"]
    .sum()
    .sort_values(by="Amount", ascending=False)
)
sns.barplot(data=sales_marital, x="Marital_Status", y="Amount", hue="Gender")
```

<h4><b>Occupation</b></h4>


```python
sns.set(rc={"figure.figsize": (20, 5)})
ax = sns.countplot(x="Occupation", data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

<h4><b>Product Category</b></h4>


```python
sns.set(rc={"figure.figsize": (27, 12)})
ax = sns.countplot(x="Product_Category", data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={"figure.figsize": (35, 15)})
sales_cate = (
    df.groupby(["Product_Category"], as_index=False)["Amount"]
    .sum()
    .sort_values(by="Amount", ascending=False)
)
sns.barplot(data=sales_cate, x="Product_Category", y="Amount")
```

```python
sns.set(rc={"figure.figsize": (20, 5)})
sales_cate = (
    df.groupby(["Product_ID"], as_index=False)["Orders"]
    .sum()
    .sort_values(by="Orders", ascending=False)
    .head(10)
)
sns.barplot(data=sales_cate, x="Product_ID", y="Orders")
```
